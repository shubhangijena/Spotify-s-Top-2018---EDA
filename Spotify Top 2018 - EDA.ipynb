```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
##### Download using conda install -c conda-forge wordcloud

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(background_color='black',).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    plt.imshow(wordcloud)
    plt.show()
```

```python
filtered_class = df[df['artists'] == 'Drake']
sns.lmplot("danceability", "valence", data=filtered_class, hue="energy", fit_reg=False, col='artists', col_wrap=2) #comparing valence with danceability
```

```python
corr = df.loc[:,df.dtypes == 'float64'].corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap=sns.diverging_palette(220, 10, as_cmap=True))
```

```python
# Melt DataFrame
df.drop(['id'], axis = 1)
melted_df = pd.melt(df, 
                    id_vars=["mode", "key", "acousticness"], # Variables to keep
                    var_name="ID") # Name of melted variable
melted_df.head()
```

```python
sns.swarmplot(x='mode', y='acousticness', data=melted_df, 
              hue='key')
```
